```python
import pandas as pd
import numpy as np
```

```python
#creating an multi-indexed dataframe

values=np.random.rand(100).reshape(10,10)

mi_rows=pd.MultiIndex.from_product([[1,2],['a','b','c','d','e']])

mi_cols=pd.MultiIndex.from_product([[3,4],['a','b','c','d','e']])
df=pd.DataFrame(values,index=mi_rows,columns=mi_cols)
df
```

```python
#creating a vanilla pd.DataFrane
vanilla=pd.DataFrame(np.random.rand(100).reshape(10,10))
vanilla.loc[1,:]
```

```python
#slicing over the first level of the index
df.loc[1]
```

```python
# slicing over the two levels of the index
df.loc[(1,'a'),:]
```

```python
# however, a naive slice attemp over the second level does not work
try:
    df.loc['a',:]
except KeyError as e:
    print('KeyError message: {}'.format(e))
```

```python
#three possibilities: 
#- swapping levels
df.swaplevel(0,1)
df.swaplevel(0,1).loc['a']
```

```python
#- adding an empty slice for the first level
df.loc[(slice(None),'a'),:]
```

```python
#(slice is necessary, using only None will raise an error)
try:
    df.loc[(None,'a'),:]
except KeyError as e:
    print('KeyError message: {}'.format(e))

df.loc[(slice(None),('a','b')),:]
```

```python
#creating a third df with a third level in its index
df3=df.set_index(np.arange(10),append=True)
df3
```

```python
# using xs, more verbose, and can access one or more level easily. 
# However, complicated slices such as df.loc[([1,2],['a','b']),:]
# do not work using xs
df.xs('a',level=1,axis=0)
df.xs('a',level=1,axis=1)
```

```python
#xs can slice several levels:
df.xs([1,'a'],level=[0,1],axis=0)
```

```python
# but cannot do stuff like:
df.loc[([1,2],['a','b']),:]
```

```python
#for xs, we can use level names instead of level index for the level argument:
df_named=df.rename_axis(['foo','bar'],axis=0)
df_named=df_named.rename_axis(['one','two'],axis=1)
df_named
```

```python
df_named.xs(1,level='foo',axis=0)
#useful drop_level arguments allows to keep the sliced level in the output:
df_named.xs(1,level='foo',axis=0,drop_level=False)
```

```python
#iloc slices: using integer position to slice
#gets the forth row
df.iloc[3,:]
```

```python
#gets the first 3 rows
df.iloc[:3,:]
```

```python
# gets the second and 4th row
df.iloc[1:5:2,:]
```

```python
# gets rows 5 to 1 in reverse order
df.iloc[5:1:-1,:]
```

```python
#getting level values of an index level:
df.index.get_level_values(0)
```

```python
##USING TIMESERIES
#we first create a daterange used for a future index
dr=pd.date_range(start='20170101',freq='1D',periods=10)
dr
```

```python
ts=pd.Series(range(10),dr)
ts=ts.to_frame('foo')
ts
```

```python
#datetime indexes are cool because we can gather time features in 
# a vectorized way:
ts2=ts.reset_index().rename(columns={'index':'TkDt'})
```

```python
# this wont work:
ts2.TkDt.day
```

```python
#this will work, but is time intensive 
ts2.TkDt.apply(lambda x: x.day)
```

```python
# this will work and is faster and less verbose
ts.index.day
```
